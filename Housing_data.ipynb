```python
ds=ds.drop(columns=['Id'])
```

```python
ds.describe()
```

```python
ds.columns  #checking the columns of the file
```

```python
ds["HouseAge"] = 2025 - ds["YearBuilt"]
ds["PricePerSqFt"] = ds["SalePrice"] / ds["GrLivArea"]
```

```python
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error



```

```python
x=ds.drop(columns=['SalePrice'])
y=ds['SalePrice']

print(x,x.shape)
print(y,y.shape)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression, Ridge,Lasso
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
print("Mean Square Error:",mean_squared_error(y_test,y_pred))
print("R2_Score:",r2_score(y_test,y_pred))
print("Mean Absolure error:",mean_absolute_error(y_test,y_pred))

```

```python
Ridge=Ridge()
Ridge.fit(x_train,y_train)
y_pred_Ridge=Ridge.predict(x_test)
print("Mean Square Error:",mean_squared_error(y_test,y_pred_Ridge))
print("R2_Score:",r2_score(y_test,y_pred_Ridge))
```

```python
lasso=Lasso()
lasso.fit(x_train,y_train)
y_pred_lasso=lasso.predict(x_test)
print("Mean Square Error:",mean_squared_error(y_test,y_pred_lasso))
print("R2_Score:",r2_score(y_test,y_pred_lasso))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
rf.fit(x_train,y_train)
y_pred_rf=rf.predict(x_test)
print("Mean Square Error:",mean_squared_error(y_test,y_pred_rf))
print("R2_Score:",r2_score(y_test,y_pred_rf))
```

```python



feature_importance = pd.Series(
    rf.feature_importances_,
    index=x.columns
).sort_values(ascending=False)

print(feature_importance)

feature_importance.plot(kind='bar',figsize=(10,6))
plt.title("Feature Importance")
plt.show()
```

```python
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x_train)
x_test_poly=poly.transform(x_test)

poly_lr=LinearRegression()
poly_lr.fit(x_poly,y_train)
y_pred_poly=poly_lr.predict(x_test_poly)
print("Mean Square Error:",mean_squared_error(y_test,y_pred_poly))
print("R2_Score:",r2_score(y_test,y_pred_poly))
```
